## 1. Tải dữ liệu

````
!wget --no-check-certificate \
https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip \
-O ./cats_and_dogs_filtered.zip
````

````
!unzip -q cats_and_dogs_filtered.zip
````

````
train_folder = './cats_and_dogs_filtered/train'
valid_folder = './cats_and_dogs_filtered/validation'
````

## 2. Load dữ liệu

````
from tensorflow.keras.preprocessing.image import ImageDataGenerator
````

### 2.1. Chuẩn hóa dữ liệu

Hàm ImageDataGenerator bổ sung một số hàm tăng cường dữ liệu sẽ được trình bày vào buổi sau.

````
train_generator = ImageDataGenerator(
    rescale=1/255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
    )
valid_generator = ImageDataGenerator(rescale=1/255)
````

### 2.2. Load nhanh dữ liệu dưới dạng generator

````
generated_train_data = train_generator.flow_from_directory(
    train_folder,
    target_size= (150, 150),
    batch_size=64,
    class_mode='binary'
)
````

````
generated_train_data.class_indices
````

````
generated_valid_data = valid_generator.flow_from_directory(
    valid_folder,
    target_size= (150, 150),
    batch_size=64,
    class_mode='binary'
)
````

## 3. Xây dựng mô hình

### 3.1. Xây dựng các lớp Convolution

````
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
````

#### TODO 1: Bạn hãy tự thiết kế các lớp Conv để trích xuất đặc trưng

````
model = Sequential(
  [Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
  MaxPooling2D(2, 2),
  Conv2D(64, (3, 3), activation='relu'),
  MaxPooling2D(2, 2),
  Conv2D(128, (3, 3), activation='relu'),
  MaxPooling2D(2, 2),
  Conv2D(128, (3, 3), activation='relu'),
  MaxPooling2D(2, 2),
]
  )
````

````
model.summary()
````

### 3.2. Thêm lớp phân loại

#### TODO 2: Bạn hãy tự thiết kế mạng nơ ron để phân loại

````
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
````

````
model.summary()
````

## 4.Tiến hành training

#### TODO 3: Thiết lập quy trình đào tạo
- Bạn hãy tự thiết kế hàm đào tạo sử dụng kèm tốc độ học
- Số lượng epochs đào tạo

````
from tensorflow.keras.optimizers import RMSprop, Adam, SGD
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['acc'])
````

Hàm theo dõi độ chính xác trên bộ validation. Hàm này sẽ dừng quá trình đào tạo khi độ chính xác trên tập này giảm 5 lần liên tục.

````
early_stopping_callback = tf.keras.callbacks.EarlyStopping(patience=10, monitor='val_acc')
````

````
history = model.fit(
    generated_train_data,
    epochs = 50,
    verbose=1,
    validation_data = generated_valid_data,
    callbacks=[early_stopping_callback]
)
````

**Độ chính xác kỳ vọng:** 79-81%

````
model.evaluate(generated_valid_data)
````

## 5. Lưu mô hình

````
model.save('cnn.h5')
````
