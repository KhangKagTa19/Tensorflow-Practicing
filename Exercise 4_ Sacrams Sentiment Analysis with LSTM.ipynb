## 1. Tải dữ liệu

````python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
````

````python
tokenizer = Tokenizer(oov_token="<OOV>")
````

````python
!wget --no-check-certificate \
    https://storage.googleapis.com/learning-datasets/sarcasm.json \
    -O /tmp/sarcasm.json
````

## 2. Xử lý dữ liệu

````python
import json

with open("/tmp/sarcasm.json", 'r') as f:
  data = json.load(f)
````

````python
for item in data[:10]:
  print(item)
````

````python
sentences = []
labels = []
for item in data:
  sentences.append(item['headline'])
  labels.append(item['is_sarcastic'])
````

````python
sentences[1]
````

````python
labels[1]
````

````python
tokenizer.fit_on_texts(sentences)
````

````python
word_index = tokenizer.word_index
````

````python
word_index
````

````python
len(word_index)
````

````python
sequences = tokenizer.texts_to_sequences(sentences)
````

````python
len(sequences[0])
````

````python
len(sequences[3])
````

````python
padded_sequences = pad_sequences(sequences, padding="pre")
````

````python
padded_sequences[0]
````

````python
train_len = int(padded_sequences.shape[0] * 0.8)
````

````python
train_len
````

````python
padded_sequences.shape
````

````python
padded_train_sequences = padded_sequences[:train_len, :]
padded_val_sequences = padded_sequences[train_len:, :]
````

````python
padded_train_sequences.shape, padded_val_sequences.shape
````

````python
padded_train_sequences[0]
````

````python
import numpy as np
import tensorflow as tf
````

````python
train_labels = tf.constant(labels[:train_len])
val_labels = tf.constant(labels[train_len:])
````

````python
padded_val_sequences.shape, val_labels.shape
````

## 3. Xây dựng mô hình
### 3.1 Mạng LSTM

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, RNN, Dense, LSTM, Bidirectional, SimpleRNN
import tensorflow as tf
````

````python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index) + 1, 300))
model.add(LSTM(32, return_sequences=True))
model.add(LSTM(32))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
````

````python
model.summary()
````

````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
````

````python
# Tạo thư mục lưu model
checkpointFolder = './tmp/checkpoints'
!os.makedirs(checkpointFolder, exist_ok=True)
````

````python
# Cài đặt lưu model tốt nhất
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpointFolder +'/model.h5',save_best_only=True,monitor='val_acc')

# Cài đặt dừng sớm khi độ chính xác mô hình giảm liên tục trên tập validation
early_stopping_callback = tf.keras.callbacks.EarlyStopping(patience=5,monitor='val_acc')
````

````python
history = model.fit(padded_train_sequences, train_labels, epochs=10, validation_data=(padded_val_sequences, val_labels),
                    callbacks=[
                        model_checkpoint_callback,
                        early_stopping_callback])
````

### 3.2 Mạng Bidirectional LSTM

Xem [video phần này](https://youtu.be/X6bYTZJkEDQ?list=PLMm4sOMuA2QI5x_0KlNT3LuKDi6-ByboB) và thực hành ứng dụng lập trình mô hình

````python
# TODO 1: Tạo model
bidirectional_model = Sequential()

# TODO 2: Thêm Embedding
bidirectional_model.add(Embedding(len(tokenizer.word_index) + 1, 64))

# TODO 3: Tạo Bidirectional với nhân LSTM
bidirectional_model.add(Bidirectional(LSTM(32)))

# TODO 4: Đưa qua mạng Dense
bidirectional_model.add(Dense(32, activation='relu'))
bidirectional_model.add(Dense(1, activation='sigmoid'))
````

````python
# TODO 5: Set up thuật toán
bidirectional_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
````

````python
# Tạo thư mục lưu model
bidirect_checkpointFolder = './tmp/bidirect_checkpoints'
!os.makedirs(bidirect_checkpointFolder, exist_ok=True)
````

````python
# TODO 6: Set up lưu checkpoint, dừng sớm và tiến hành đào tạo
bidirectional_model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath=bidirect_checkpointFolder +'/bidirect_model.h5',save_best_only=True,monitor='val_acc')


bidirectional_early_stopping_callback = tf.keras.callbacks.EarlyStopping(patience=5,monitor='val_acc')

history = bidirectional_model.fit(padded_train_sequences, train_labels, epochs=10, validation_data=(padded_val_sequences, val_labels),
                    callbacks=[
                        bidirectional_model_checkpoint_callback,
                        bidirectional_early_stopping_callback])
````

**Yêu cầu kết quả**:

Độ chính xác trên tập validation: **84%-86%**

### Mạng Deep Bidirectional

````python
deep_bidirectional_model = Sequential()
deep_bidirectional_model.add(Embedding(len(tokenizer.word_index) + 1, 256))
# TODO 7: Tạo mạng deep Bidirectional với số lượng lớp sâu tùy biến
deep_bidirectional_model.add(Bidirectional(LSTM(64, return_sequences=True)))
deep_bidirectional_model.add(Bidirectional(LSTM(32)))
deep_bidirectional_model.add(Dense(32, activation='relu'))
deep_bidirectional_model.add(Dense(1, activation='sigmoid'))
````

````python
# TODO 8: Setup training để đạt được mục tiêu

deep_bidirectional_model.compile(optimizer=tf.keras.optimizers.Adam(), loss='binary_crossentropy', metrics=['acc'])
````

````python
# Tạo thư mục lưu model
deep_bidirect_checkpointFolder = './tmp/deep_bidirect_checkpoints'
!os.makedirs(deep_bidirect_checkpointFolder, exist_ok=True)
````

````python
deep_bidirectional_model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath=deep_bidirect_checkpointFolder +'/deep_bi_model.h5',save_best_only=True,monitor='val_acc')


early_stopping_callback = tf.keras.callbacks.EarlyStopping(patience=5,monitor='val_acc')

history = deep_bidirectional_model.fit(padded_train_sequences, train_labels, epochs=100, validation_data=(padded_val_sequences, val_labels),
                    callbacks=[
                        deep_bidirectional_model_checkpoint_callback,
                        early_stopping_callback])
````

**Yêu cầu kết quả**:

Độ chính xác trên tập validation: **86%-87%**
